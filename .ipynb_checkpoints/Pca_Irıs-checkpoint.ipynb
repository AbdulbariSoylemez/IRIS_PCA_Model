```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
data=pd.read_csv("IRIS.csv")
data.head(10)
```

```python
degiskenler = ['çanak_uzunluğu', 'çanak_genişliği', 'taç_uzunluğu', 'taç_genişliği', 'tür']
data.columns=degiskenler
```

```python
x=data.drop(["tür"],axis=1)
y=data[["tür"]]
```

```python
x=StandardScaler().fit_transform(x)
```

```python
x
```

```python
pca=PCA(n_components=2)
principalComponents=pca.fit_transform(x)
YeniİkiBoyut=pd.DataFrame(data=principalComponents,columns=["AnaBileşen 1","AnaBileşen 2"])
```

```python
yenidata = pd.concat([YeniİkiBoyut, data[["tür"]]], axis=1)
```

```python
türler=["Iris-setosa","Iris-versicolor","Iris-virginica"]
renk=["g","b","r"]
for tü,re in zip(türler,renk):
    dftemp=yenidata[data["tür"]==tü]
    plt.scatter(dftemp["AnaBileşen 1"],dftemp["AnaBileşen 2"],color=re)
```

```python
pca.explained_variance_ratio_
```
